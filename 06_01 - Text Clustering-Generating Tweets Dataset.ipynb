We have a lot of re-tweets. Let's define a function that generates a dataset of a specific size without retweets

```python
def generate_dataset(query, size):
    count = 0
    index = 1
    next_max_id = ''
    tries = 1
    dataset = set() #Helps in generating unique tweets 
    max_tries = 10
    
    search_result=twitter.search(q=query, lang='en', count=100, max_id=next_max_id)
    while (count<size):
        adding = False
        for result in search_result['statuses']:
            text = result['text']
            if not text.startswith('RT') and not text in dataset:
                dataset.add(text)
                print '[%s] %s' % (count, text)
                adding = True
                count+=1
                if (count==size):
                    break
        next_max_id = search_result['search_metadata']['next_results'].split('max_id=')[1].split('&')[0]
        result = twitter.search(q=query, lang='en', count=100, max_id=next_max_id)

        if (not adding):
            tries+=1
            
        if (tries>max_tries):
            break

    return dataset

dataset = generate_dataset("soda",1000)
```

Save the dataset in a file

```python
path = 'C:\\tmp\\tweets.txt'

text_file = open(path, "a")
for line in dataset:
    text_file.write(line.encode('ascii','ignore'))
text_file.close()
```
